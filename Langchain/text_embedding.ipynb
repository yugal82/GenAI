#### Huggingface embedding technique

```python
os.environ["HUGGING_FACE_API_KEY"] = os.getenv("HUGGING_FACE_API_KEY")
```

```python

from langchain_huggingface import HuggingFaceEmbeddings
hf_embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "This is a test document"
query_result = hf_embeddings.embed_query(text)
print(query_result)
print(len(query_result))
```

```python
# we can also do it for list of documents
texts = ["This is a test document", "This is another test document"]
query_result_list = hf_embeddings.embed_documents(texts)
print(query_result_list)
print(len(query_result_list))
print(len(query_result_list[0]))
```
