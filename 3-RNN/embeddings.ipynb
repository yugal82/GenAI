```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = ['the glass of milk', 'the glass of juice', 'the cup of tea', 'I am a good boy', 'I am a good developer', 'understand the meaning of words', 'your videos are good']
```

```python
vocab_size = 10000
one_hot_rep = [one_hot(words, vocab_size)for words in sentences]
one_hot_rep
```

```python
# define max sentence len
sent_len = 8
embedded_docs = pad_sequences(one_hot_rep, padding='pre', maxlen=sent_len)
embedded_docs
```

```python
# feature representation
dim = 10
model = Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_len))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
